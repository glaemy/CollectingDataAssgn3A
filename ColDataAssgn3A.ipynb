```python
#In order to find out the relationship between the rating of an anime and the number of members it has,
#I begin by isolating the two columns first
anime_df[['rating','members']]
#Then I import matplotlib to create a scatter plot of their relationship
#I chose a scatter plot because of the huge amount of data where the point of every plot makes it easier to see outliers
import matplotlib.pyplot as plt
anime_df[['rating', 'members']].plot(kind='scatter', x='rating', y='members', title='Relation Between Rating and Members')
plt.xlabel('Rating')
plt.ylabel('Members')
plt.show()
```

```python
#The plot tells us that the highest number of members are enjoyed by anime with a rating close to 8 
#while the low membership for anime with lower rating is expected, it is notable that those with an 8 have more membership than rating of 10
#The data can be interpeted as people generally enjoy moderately high rating anime more 
#Maybe because those which are too good may pose higher pressure to enjoy them
```
